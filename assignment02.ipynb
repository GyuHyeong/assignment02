```python
import numpy as np
import matplotlib.pyplot as plt
```

# define my function:  $f(x)=x^2+2x$

```python
def myFunction(x):
    f = x**2 + 2*x
    return f
```

# define the derivative of my function: $f'(x) = 2x+2$

```python
def myDerivativeFunction(x):
    Df = 2*x + 2
    return Df
```

# define the domain of the function: $x = [-10 : 0.1 : 10]$

```python
x = np.arange(-10, 10, 0.1)
```

# compute the graph

```python
f = myFunction(x)
Df = myDerivativeFunction(x)
```

# pick 3 points x=-3,0,5

```python
a= np.array([-3,0,5])
```

# plot the graphs for the function and its derivative

```python
plt.figure(1)
plt.plot(x, f, 'b', label="function")
plt.plot(x, Df, 'r', label="derivative")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

# define Taylor approximation of the function. Near x=a, $ \hat{f}(x)=f(a)+f'(a)*(x-a)$

```python
def myTaylorApproximation(x,a):
    Ta=myFunction(a)+myDerivativeFunction(a)*(x-a)
    return Ta
```

```python
Ta1=myTaylorApproximation(x,a[0])
Ta2=myTaylorApproximation(x,a[1])
Ta3=myTaylorApproximation(x,a[2])
plt.figure(1)
plt.plot(x, f, 'b', label="function")
plt.plot(x, Ta1, 'r', label="Taylor approximation 1")
plt.plot(x, Ta2, 'g', label="Taylor approximation 2")
plt.plot(x, Ta3, 'y', label="Taylor approximation 3")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
